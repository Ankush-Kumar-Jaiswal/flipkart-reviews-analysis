```python
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
import requests
import time
import random
```

```python

# Replace with the product's reviews page URL
url = 'https://www.flipkart.com/muscleblaze-biozyme-performance-whey-protein-powder-25g-per-scoop/product-reviews/itm239e397ea2575?pid=PSLGGRDAMNDGH8XQ&lid=LSTPSLGGRDAMNDGH8XQBBZUTO&marketplace=FLIPKART&page='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0 Safari/537.36'
}

reviews_data=[]
i=1
while i<148:
    response = requests.get(url+str(i), headers=headers)
    time.sleep(random.uniform(1, 5))

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # All review blocks
        reviews = soup.find_all('div', {'class': 'EKFha-'})
        for idx, review in enumerate(reviews, 1):
            # Extract the review title, rating, and text
            rating_tag = review.find('div', {'class': 'Ga3i8K'})
            rating = rating_tag.text.strip() if rating_tag else 'No rating'
            
            title_tag = review.find('p', {'class': 'z9E0IG'})
            title = title_tag.text.strip() if title_tag else 'No title'
            
            text_tag = review.find('div', {'class': 'ZmyHeo'})
            text = text_tag.text.strip() if text_tag else 'No review text'
    
            buyer_tag = review.find('p', {'class': 'MztJPv'})
            buyer = title_tag.text.strip() if title_tag else 'No title'
    
            reviews_data.append({"Review": idx, "Rating": rating, "Title": title, "Text": text, "Buyer_tag": buyer})
        
        print(f"Parsing done for page no {i}")
        i+=1
    
    else:
        time.sleep(random.uniform(4, 10))

        print(f"Failed to fetch page, status code: {response.status_code}")

```

```python
data=pd.DataFrame(reviews_data)
data.shape
```

```python
data.to_csv("muscleblaze_review.csv", index=False)
```
